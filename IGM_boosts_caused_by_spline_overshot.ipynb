```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
with_igm = pd.read_csv("data/with_spectra.txt")
without_igm = pd.read_csv("data/without_spectra.txt")
diff = without_igm["y"]-with_igm["y"]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.axvspan(8839.72, 10933.2, alpha=0.1, color='red')

plt.plot(with_igm["x"],with_igm["y"])
plt.plot(without_igm["x"],without_igm["y"])
plt.xlim(0.0,12000.0)
```

```python
boost_mask = with_igm["y"]>without_igm["y"]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.axvspan(8839.72, 10933.2, alpha=0.1, color='red')

plt.loglog(with_igm["x"],with_igm["y"])
plt.loglog(without_igm["x"],without_igm["y"])
```

```python
with_igm_spline_interp = pd.read_csv("data/with_poly.txt")
without_igm_spline_interp = pd.read_csv("data/without_poly.txt")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.axvspan(8839.72, 10933.2, alpha=0.1, color='red')

plt.title("Interpolated Spline with knot points")
plt.loglog(with_igm_spline_interp["x"],with_igm_spline_interp["y"])
plt.loglog(without_igm_spline_interp["x"],without_igm_spline_interp["y"])
plt.loglog(with_igm["x"],with_igm["y"],'X',color='green')
plt.xlim(8800, 11000)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.axvline(8839.72, alpha=0.1, color='red')

plt.title("Spline with knot point")
plt.annotate(xy=[9e3,3e39], s="Knot Point")
plt.loglog(with_igm_spline_interp["x"],with_igm_spline_interp["y"])
plt.loglog(without_igm_spline_interp["x"],without_igm_spline_interp["y"])
plt.loglog(with_igm["x"],with_igm["y"],'X',color='green')
plt.xlim(with_igm_spline_interp["x"][0]-10, with_igm_spline_interp["x"][100])
plt.ylim(5e36,6e39)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.axvline(8839.72, alpha=0.1, color='red')

plt.title("Spline with knot point. Note the overshoot.")
plt.annotate(xy=[9e3,4e39], s="Knot Point")
plt.plot(with_igm_spline_interp["x"],with_igm_spline_interp["y"],'.')
plt.plot(without_igm_spline_interp["x"],without_igm_spline_interp["y"])
plt.plot(with_igm["x"],with_igm["y"],'X',color='green')
plt.xlim(with_igm_spline_interp["x"][0]-10, with_igm_spline_interp["x"][100])
plt.ylim(5e36,6e39)
```
